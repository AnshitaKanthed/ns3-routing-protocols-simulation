```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the data
dsdv_df = pd.read_csv('data/DSDV.csv')
dsr_df = pd.read_csv('data/DSR.csv')
aodv_df = pd.read_csv('data/AODV.csv')
```

```python
# Calculating Packet Delivery Fraction for each protocol
dsdv_df['PacketDeliveryFraction'] = dsdv_df['PacketsReceived'] / dsdv_df['NumberOfSinks']
dsr_df['PacketDeliveryFraction'] = dsr_df['PacketsReceived'] / dsr_df['NumberOfSinks']
aodv_df['PacketDeliveryFraction'] = aodv_df['PacketsReceived'] / aodv_df['NumberOfSinks']

# Plotting Packet Delivery Fraction
plt.figure(figsize=(10, 6))
plt.plot(dsdv_df['SimulationSecond'], dsdv_df['PacketDeliveryFraction'], label='DSDV')
plt.plot(dsr_df['SimulationSecond'], dsr_df['PacketDeliveryFraction'], label='DSR')
plt.plot(aodv_df['SimulationSecond'], aodv_df['PacketDeliveryFraction'], label='AODV')
plt.xlabel('Simulation Second')
plt.ylabel('Packet Delivery Fraction')
plt.title('Packet Delivery Fraction Comparison')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
# Calculating Throughput (packets/second)
dsdv_df['Throughput'] = dsdv_df['PacketsReceived'] / dsdv_df['SimulationSecond']
dsr_df['Throughput'] = dsr_df['PacketsReceived'] / dsr_df['SimulationSecond']
aodv_df['Throughput'] = aodv_df['PacketsReceived'] / aodv_df['SimulationSecond']


# Plotting Throughput (bar graph)
plt.figure(figsize=(10, 6))
plt.bar(dsdv_df['SimulationSecond'], dsdv_df['Throughput'], label='DSDV', width=0.2, align='center')
plt.bar(dsr_df['SimulationSecond'] + 0.4, dsr_df['Throughput'], label='DSR', width=0.2, align='center')
plt.bar(aodv_df['SimulationSecond'] + 0.8, aodv_df['Throughput'], label='AODV', width=0.2, align='center')
plt.xlabel('Simulation Second')
plt.ylabel('Throughput (packets/second)')
plt.title('Throughput Comparison')
plt.legend()
plt.grid(axis='y')
plt.show()
```
