```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
aodv_data = pd.read_csv("data/aodv-sim2.csv")
dsdv_data = pd.read_csv("data/dsdv-sim2.csv")
dsr_data = pd.read_csv("data/dsr-sim2.csv")
```

```python
plt.figure(figsize=(10, 6))

plt.plot(aodv_data['PACKETSTRANSMITTED'], aodv_data['RTTMEDIAN'], marker='o', color='b', label='AODV')
plt.plot(dsdv_data['PACKETSTRANSMITTED'], dsdv_data['RTTMEDIAN'], marker='o', color='r', label='DSDV')
plt.plot(dsr_data['PACKETSTRANSMITTED'], dsr_data['RTTMEDIAN'], marker='o', color='g', label='DSR')

plt.xlabel('Packets')
plt.ylabel('RTT Median (ms)')
plt.title('RTT Median Comparison')
plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(aodv_data['PACKETSTRANSMITTED'], aodv_data['PACKETLOSS'], marker='o', color='b', label='AODV')
plt.plot(dsdv_data['PACKETSTRANSMITTED'], dsdv_data['PACKETLOSS'], marker='o', color='r', label='DSDV')
plt.plot(dsr_data['PACKETSTRANSMITTED'], dsr_data['PACKETLOSS'], marker='o', color='g', label='DSR')

plt.xlabel('Packets')
plt.ylabel('Packet Losses')
plt.title('Packet Losses for Each Routing Protocol')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Calculate packet loss rate
aodv_data['PACKETLOSSRATE'] = (aodv_data['PACKETLOSS'] / aodv_data['PACKETSTRANSMITTED']) * 100
dsdv_data['PACKETLOSSRATE'] = (dsdv_data['PACKETLOSS'] / dsdv_data['PACKETSTRANSMITTED']) * 100
dsr_data['PACKETLOSSRATE'] = (dsr_data['PACKETLOSS'] / dsr_data['PACKETSTRANSMITTED']) * 100


plt.figure(figsize=(10, 6))
plt.plot(aodv_data['PACKETSTRANSMITTED'], aodv_data['PACKETLOSSRATE'], marker='o', color='b', label='AODV')
plt.plot(dsdv_data['PACKETSTRANSMITTED'], dsdv_data['PACKETLOSSRATE'], marker='o', color='r', label='DSDV')
plt.plot(dsr_data['PACKETSTRANSMITTED'], dsr_data['PACKETLOSSRATE'], marker='o', color='g', label='DSR')


plt.xlabel('Packets')
plt.ylabel('Packet Loss Rate (%)')
plt.title('Packet Loss Rate for Each Routing Protocol')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Calculate RTT Variability
aodv_data['RTT_VARIABILITY'] = np.sqrt(aodv_data['RTTMEDIAN'].var())
dsdv_data['RTT_VARIABILITY'] = np.sqrt(dsdv_data['RTTMEDIAN'].var())
dsr_data['RTT_VARIABILITY'] = np.sqrt(dsr_data['RTTMEDIAN'].var())


plt.figure(figsize=(10, 6))
plt.plot(aodv_data['PACKETSTRANSMITTED'], aodv_data['RTT_VARIABILITY'], marker='o', color='b', label='AODV')
plt.plot(dsdv_data['PACKETSTRANSMITTED'], dsdv_data['RTT_VARIABILITY'], marker='o', color='r', label='DSDV')
plt.plot(dsr_data['PACKETSTRANSMITTED'], dsr_data['RTT_VARIABILITY'], marker='o', color='g', label='DSR')

plt.xlabel('Packets')
plt.ylabel('RTT Variability')
plt.title('RTT Variability for Each Routing Protocol')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Calculate control message overhead
aodv_data['NETWORK_OVERHEAD'] = (aodv_data['PACKETSTRANSMITTED'] - aodv_data['PACKETSRECEIVED']) / aodv_data['PACKETSTRANSMITTED']
dsdv_data['NETWORK_OVERHEAD'] = (dsdv_data['PACKETSTRANSMITTED'] - dsdv_data['PACKETSRECEIVED']) / dsdv_data['PACKETSTRANSMITTED']
dsr_data['NETWORK_OVERHEAD'] = (dsr_data['PACKETSTRANSMITTED'] - dsr_data['PACKETSRECEIVED']) / dsr_data['PACKETSTRANSMITTED']

plt.figure(figsize=(10, 6))
plt.plot(aodv_data['PACKETSTRANSMITTED'], aodv_data['NETWORK_OVERHEAD'], marker='o', color='b', label='AODV')
plt.plot(dsdv_data['PACKETSTRANSMITTED'], dsdv_data['NETWORK_OVERHEAD'], marker='o', color='r', label='DSDV')
plt.plot(dsr_data['PACKETSTRANSMITTED'], dsr_data['NETWORK_OVERHEAD'], marker='o', color='g', label='DSR')


plt.xlabel('Packets')
plt.ylabel('Network Overhead')
plt.title('Network Overhead for Each Routing Protocol')
plt.legend()
plt.grid(True)

plt.show()
```
